```python
# Python get today's date
# import datetime
from datetime import date
today =date.today()
print('current year', today.year)
print('current month', today.month)
print('current date', today.day)
```

```python
# Current date in different formats
# from datetime import date

today = date.today()

# dd/mm/YY
d1 = today.strftime("%d/%m/%Y")
print("d1 =", d1)

# Textual month, day and year	
d2 = today.strftime("%B %d, %Y")
print("d2 =", d2)

# mm/dd/y
d3 = today.strftime("%m/%d/%y")
print("d3 =", d3)

# Month abbreviation, day and year	
d4 = today.strftime("%b-%d-%Y")
print("d4 =", d4)
```

```python
# Get the current date and time

from datetime import datetime

# datetime object containing current date and time
now = datetime.now()
print("now =", now)

# dd/mm/YY H:M:S
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("date and time =", dt_string)
```

```python
# Difference between two dates

from datetime import datetime,date
  
# datetime(year, month, day, hour, minute, second) 
a = datetime(year=2020,month= 8,day=16) 
b = datetime(year=2019,month= 8,day=16) 
  
# returns a timedelta object 
c = a-b  
print('c = ', c) 
```

```python
from datetime import datetime,date
  
#datetime(year, month, day, hour, minute, second) 
d = datetime(year=2020,month= 8,day=16,hour=8,minute=45,second=10) 
e = datetime(year=2019,month= 8,day=16,hour=10,minute=45,second=10) 
  
# returns a timedelta object 
f = d-e  
print('f = ', f) 
```
